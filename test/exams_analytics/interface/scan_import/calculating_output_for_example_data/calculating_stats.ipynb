Let's calculate how the DDBB should look like with the example_data.xml file

```python
import xml.etree.ElementTree as ET

# Replace 'your_file.xml' with the path to your XML file
tree = ET.parse("/workspace/test/exams_analytics/interface/scan_import/example_data.xml")
root = tree.getroot()

marks_per_student_test = {}

# Print the root element and its children
for mcq_result in root.findall('mcq-test-result'):
            student_number : str = mcq_result.find('student-number').text    #type: ignore
            test_id = mcq_result.find('test-id').text                       #type: ignore
            available = int(mcq_result.find('summary-marks').get('available'))   #type: ignore
            obtained = int(mcq_result.find('summary-marks').get('obtained'))     #type: ignore
            id = f"{student_number}_{test_id}"
            if id not in marks_per_student_test:
                marks_per_student_test[id] = (available, obtained, student_number, test_id)
            else:
                if marks_per_student_test[id][0] < available:
                    marks_per_student_test[id] = (available, obtained, student_number, test_id)
                elif marks_per_student_test[id][0] == available:
                    marks_per_student_test[id] = (available, max(marks_per_student_test[id][1], obtained), student_number, test_id)

for m in marks_per_student_test.values():
    print(f"{m[2]}, {m[3]}, {m[0]}, {m[1]}")

#let's put it in a csv file
import csv 
# Specify the CSV file to write to
csv_file_path = "/workspace/test/exams_analytics/interface/scan_import/student_marks.csv"

# Write the data to a CSV file
with open(csv_file_path, mode='w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    # Write header
    csv_writer.writerow(['Student Number', 'Test ID', 'Available Marks', 'Obtained Marks'])

    # Write the student marks data
    for m in marks_per_student_test.values():
        csv_writer.writerow([m[2], m[3], m[0], m[1]])
```
